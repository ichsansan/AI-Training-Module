# Pengenalan Datasets

Di subbab ini kita akan mengenal beberapa dataset terbuka yang tersedia dan dapat digunakan secara bebas. Berikut adalah beberapa sumber yang dapat Anda gunakan untuk menemukan dataset:

* **[Scikit-Learn](https://scikit-learn.org/stable/datasets/toy_dataset.html)**: Scikit-Learn menyediakan sejumlah dataset bawaan yang sangat berguna untuk pemula. Anda dapat mengimpor dataset ini langsung dalam kode Python Anda. Contoh dataset yang dapat Anda gunakan adalah Iris, Wine, dan Breast Cancer.

* **[Kaggle](https://www.kaggle.com/datasets)**: Kaggle adalah platform online yang menyediakan akses ke berbagai dataset dan kompetisi dalam ilmu data dan kecerdasan buatan. * **[UCI Machine Learning Repository](https://archive.ics.uci.edu/datasets)**: Universitas California, Irvine (UCI) menyediakan repository dataset machine learning yang beragam. Mereka memiliki berbagai dataset klasifikasi yang dapat diunduh secara gratis.

# Visualisasi dataset

Visualisasi data membantu dalam mengidentifikasi pola, tren, anomali, dan wawasan yang mungkin sulit ditemukan dalam data mentah. Ada beberapa jenis library yang dapat digunakan untuk visualisasi data.

## 1. Matplotlib

```python
import matplotlib.pyplot as plt

x = pd.to_datetime(data_csv['Timestamp'])
y = data_csv['Gross Load']

plt.figure(figsize=(16, 5))
plt.plot(x, y)
plt.title('Gross Load')
plt.show()
```

## 2. Plotly

```python
import plotly.express as px

fig = px.scatter(data_breast_cancer, x='mean radius', y='mean texture', color='target')
fig.update_layout(height=600)
fig.show()
```

## 3. Seaborn

```python
import seaborn as sns

sns.catplot(
    data=data_iris[['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']].melt(), 
    x="value", y="variable", kind="violin"
)
```
